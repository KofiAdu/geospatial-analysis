```python
import folium
import numpy as np
import pandas as pd
from folium.plugins import HeatMap, HeatMapWithTime
```

```python
#read data
df = pd.read_csv('air_quality_los_angeles_data_2024.csv')
```

```python
df.head(10)
```

```python
#convert data column to date/time format to extract week number
#this for to measure the air quality per week
#extract date and do conversion
df["Date"] = pd.to_datetime(df["Date"])

#add new week number column
df["Week"] = df["Date"].dt.isocalendar().week
```

```python
df.head(10)
```

```python
#using numpy to calc the weekly mean Air Quality Index
weekly_average_aqi = df.groupby(["Local Site Name", "Week"]).agg({
    #get latitude value for each site from the df
    'Site Latitude': 'first',
    'Site Longitude': 'first',
    'Daily AQI Value': lambda x: np.mean(x)
}).reset_index()

weekly_average_aqi.head()
```

```python
#print("Weekly Mean AQI Value per location: ")
```

```python
#iteratte through the locations
for location in weekly_average_aqi["Local Site Name"].unique():
    location_data = weekly_average_aqi[weekly_average_aqi["Local Site Name"] == location]
    print(f"Location: {location}")
    for index, row in location_data.iterrows():
        print(f"Week: {row['Week']}")
        print(f"Average AQI Value: {row['Daily AQI Value']:.2f}")
    print('\n')
```

```python
#no geocoding needed since the data already has lat and lon
#create list to store lat, lon and aqi value
df_weekly_list = []

#iterate through aqi values for each week (52 in this case)
for week in range(1, 53):
    weekly_data = weekly_average_aqi[weekly_average_aqi["Week"] == week] [['Site Latitude', 'Site Longitude', 'Daily AQI Value']]
    df_weekly_list.append(weekly_data.values.tolist())
```

```python
#create basemap
#remeber to use a negativ esign if the longitue is West , lik 20.8976 W
base_map = folium.Map(location=[34.0549, -118.2426], zoom_start=12)
```

```python
#time heat map
HeatMapWithTime(df_weekly_list, radius=70, gradient={0.05:'blue', 0.5:'green', 0.75:'yellow', 1.0:'red'}, min_opacity=0.5, max_opacity=0.8, use_local_extrema=True).add_to(base_map)
```

```python
base_map.save("weekly_air_quality_index.html")
```
